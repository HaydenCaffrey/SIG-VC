```python
import glob 
import numpy as np 
import tqdm 
import random
from IPython.display import Audio
import librosa
import os
import re
import matplotlib.pyplot as plt
```

```python
GTABASE = '/Netdata/2020/TTS/ljs_db1_db4_blm_vocoder/mels_gta'
```

```python
gtas = [n for n in glob.glob(os.path.join(GTABASE, '*'))]
```

```python
random.sample(gtas, 10)
```

```python
name_pattern = re.compile(r'mel\-(?P<dataset>[a-z0-9]+)_(?P<sid>[_a-zA-Z0-9\-\.]+)\.wav\.npy')
def parse_name(name) : 
    m = name_pattern.match(name)
    assert m is not None, f'invalid syntax : name : {name}'
    groups = m.groupdict()
    return (groups['dataset'], groups['sid'])
```

```python
# db1, db4, ljs, blm
def find_db1(sid : str) -> str : 
    # assume db-1 means db1 CN
    return os.path.join(
        '/NASdata/AudioData/mandarin/合成数据/中文女声DB-1/CN/Wave',
        f'CN{sid}.wav'
    )

def find_db4(sid : str) -> str : 
    if sid[0] in {'E', 'C'} : 
        lang = sid[:2]
        lsid = sid[2:]
    else : 
        lang = sid[:3]
        lsid = sid[3:]
    return os.path.join(
        '/NASdata/AudioData/mandarin/合成数据/中文女生DB-4',
        lang,
        'wav',
        f'{lsid}.wav'
    )

blm_cate_to_dir = {
    'an' : '/NASdata/AudioData/mandarin/合成数据/male_for_arrest-record/answers/16k_mono_vad',
    'cont' : '/NASdata/AudioData/mandarin/合成数据/male_for_arrest-record/contents/16k_mono_vad',
    'qu' : '/NASdata/AudioData/mandarin/合成数据/male_for_arrest-record/questions/16k_mono_vad'
}
def find_blm(sid : str) -> str : 
    parts = sid.split('_')
    assert len(parts) == 2, f'unexpected blm sid syntax : {}'
    cate = parts[0]
    csid = parts[1]
    
    return os.path.join(
        blm_cate_to_dir[cate],
        f'{cate}_{csid}.wav'
    )

def find_ljs(sid : str) -> str :
    return os.path.join(
        '/NASdata/AudioData/english/LJSpeech-1.1/wavs',
        f'{sid}.wav'
    )

def find_path(parsed) : 
    data, sid = parsed
    if data == 'db1' : 
        return find_db1(sid)
    elif data == 'db4' : 
        return find_db4(sid)
    elif data == 'ljs' : 
        return find_ljs(sid)
    else : 
        return None
```

```python
names = [os.path.basename(n) for n in random.sample(gtas, 10)]
parsed = [parse_name(name) for name in names]
for p,n in zip(parsed, names) : 
    print(f'{p}\t{n}\t{find_path(p)}')
```

```python
def show_data(path) : 
    basename = os.path.basename(path)
    parsed = parse_name(basename)
    audio_path = find_path(parsed)
    
    assert audio_path is not None, f'path : {path} is bad'
    
    mel = np.load(path)
    audio, sr = librosa.load(audio_path, sr=16000)
    print(f'loaded sample rate : {sr}')
    print(mel.shape)
    display(Audio(audio, rate=16000))
    plt.imshow(mel[:32,].T, cmap='coolwarm', aspect='auto')
    
```

```python
idx = 6
print(gtas[idx])
show_data(gtas[idx])
```

```python
show_data('/Netdata/2020/TTS/ljs_db1_db4_blm_vocoder/mels_gta/mel-db1_001606.wav.npy')
```

```python
def map_name_to_metaline(path) : 
    basename = os.path.basename(path)
    parsed = parse_name(basename)
    audio_path = find_path(parsed)
    if audio_path is None : 
        return None
    else :
        return '|'.join([audio_path, '<>', path])
```

```python
buffer = [l for l in [map_name_to_metaline(path) for path in gtas] if l is not None]
random.shuffle(buffer)
n_test = 100
test_buffer = buffer[:100]
train_buffer = buffer[100:]

with open('dataset/db1_db4_ljs_gta/metadata.txt', 'w') as f : 
    f.write('\n'.join(buffer))
    
with open('dataset/db1_db4_ljs_gta/train_files.txt', 'w') as f : 
    f.write('\n'.join(train_buffer))
    
with open('dataset/db1_db4_ljs_gta/test_files.txt', 'w') as f : 
    f.write('\n'.join(test_buffer))
```
